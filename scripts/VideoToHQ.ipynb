# VideoToHQ (Colab)
Este notebook contém células prontas para executar no Google Colab:
- Clonar o repositório
- Instalar dependências (`ffmpeg` + `pip install -r requirements.txt`)
- Baixar vídeo e legendas com `baixar_youtube_com_legenda.py` (renomeia para `_video.mp4`)
- Converter `_video.mp4` para `video.mp4` com `ffmpeg`
- Gerar a HQ leve com `srt_to_hq_pdf_leve.py`

Edite a célula de download e substitua `PASTE_VIDEO_URL_HERE` pela URL do YouTube antes de executar.

```python
%%bash
set -e
git clone https://github.com/SamuelSBJr97/SSBJr.HQCourses.git
cd SSBJr.HQCourses/scripts
apt-get update -qq
apt-get install -y -qq ffmpeg
pip install -r requirements.txt
```

```python
%%bash
set -e
cd SSBJr.HQCourses/scripts
# Substitua a variável VIDEO_URL pela URL desejada antes de executar
VIDEO_URL="PASTE_VIDEO_URL_HERE"
python3 baixar_youtube_com_legenda.py "$VIDEO_URL"
```

```python
%%bash
cd SSBJr.HQCourses/scripts
# O script já renomeia o arquivo baixado para _video.mp4. Aqui convertemos para video.mp4
if [ -f "_video.mp4" ]; then
  ffmpeg -y -i _video.mp4 -vf scale=1280:-2 -c:v libx264 -preset fast -crf 23 -c:a copy video.mp4
  echo "Conversão concluída: video.mp4"
else
  echo "_video.mp4 não encontrado. Execute a célula de download primeiro."
fi
```

```python
%%bash
cd SSBJr.HQCourses/scripts
# Gera o PDF leve usando as legendas (transcript.srt ou transcript.srv3)
python3 srt_to_hq_pdf_leve.py
```
